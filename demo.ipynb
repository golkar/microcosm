```python
from microcosm import sample_world
```

```python
series, param_dict = sample_world(seed = 0)
```

```python
# generated an observation of a system with 5 constituents
param_dict['num_sys']
```

```python
# Here are is the descriptions of the first subsystem
param_dict[0]
# step_multip is the step size subsampling factor
```

```python
# The sequence is concatenated and normalized using the following parameters
param_dict['normalization'] # mean and std
```

```python
# And then embedded in a different coordinate system
# param_dict['embedding'] gives the svd of linear embedding
# the second component is the dilation factor
param_dict['embedding'][1]
# In this case, it is a multiscale dilation with factors of 1 and 10.
```
